This notebook introduces Named Entity Recognition (NER) using spaCy's Spanish model.
We process a sample sentence, extract entities (persons, organisations, locations),
visualise them with displaCy, and inspect the full linguistic token analysis.

## 1. Setup: Importing Libraries

````python
import spacy
from spacy import displacy
````

## 2. Loading the Spanish NLP Model

We use `es_core_news_md` (medium Spanish model trained on news text).
If the model is not yet installed, run once:
```
python -m spacy download es_core_news_md
```

````python
nlp = spacy.load("es_core_news_md")
````

## 3. Processing a Sample Sentence

The sentence below is the NER demonstration input (Spanish text is kept intentionally
to showcase the model's multilingual capability on real-world news-style text).

````python
# Sample Spanish sentence for NER demonstration
texto = (
    "El presidente Pedro Sánchez visitó la sede de Naciones Unidas "
    "en Nueva York el 20 de septiembre."
)
````

````python
doc = nlp(texto)
````

## 4. Extracting Named Entities

````python
for ent in doc.ents:
    print(ent.text, ent.label_)
````

> **Output interpretation:** Each line shows an entity span and its label. `PER` = person (Pedro Sánchez), `ORG` = organisation (Naciones Unidas), `LOC` = location (Nueva York). This confirms the model correctly identifies the three main NER categories present in the sentence.

## 5. Visualising Entities with displaCy

````python
displacy.render(doc, style="ent", jupyter=True)
````

> **Output interpretation:** Renders the sentence inline with colour-coded entity spans. Each highlighted segment corresponds to a recognised named entity, making it easy to visually verify NER precision before integrating the pipeline into a production system.

## 6. Token-Level Linguistic Analysis

spaCy annotates every token with part-of-speech tags, dependency relations,
lemmas, and morphological features — all produced in a single pass through the pipeline.

````python
for token in doc:
    print(
        token.text,
        token.lemma_,
        token.pos_,
        token.tag_,
        token.dep_,
        token.shape_,
        token.is_alpha,
        token.is_stop,
    )
````

> **Output interpretation:** Each row corresponds to one token and shows: surface form, lemma (base form), universal POS tag, fine-grained tag, syntactic dependency label, shape pattern, whether alphabetic, and whether a stop word. This rich annotation enables downstream tasks such as lemma-based search, syntactic parsing, and feature extraction for ML models.

## 7. Dependency Tree Visualisation

The dependency parse shows grammatical relations between tokens (subject, object, modifier, etc.).
Note: `displacy.render` is the correct call inside Jupyter (instead of `displacy.serve`,
which would start a separate web server and block execution).

````python
displacy.render(doc, style="dep", jupyter=True)
````

> **Output interpretation:** The dependency tree diagram shows how every token relates to its syntactic head. The root verb `visitó` (visited) governs the subject `presidente` and the object `sede`, while proper-noun chains (Pedro→Sánchez, Naciones→Unidas) illustrate how multi-token entities are linked.
