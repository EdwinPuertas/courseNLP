## What is Named Entity Recognition (NER)?

NER is a sequence labelling task that identifies and classifies **named entities** (persons, organisations, locations, dates, etc.) in text. spaCy uses a transition-based named entity recogniser built on a neural network architecture.


---

## 1. Environment Setup


```python
import json
import random
from pathlib import Path

import spacy
from spacy.training import Example, offsets_to_biluo_tags

DIR_DATA = Path.cwd() / 'data'
```

---

## 2. Verify Entity Offset Format

spaCy requires entity annotations as character offsets: `(start, end, LABEL)`. The helper `offsets_to_biluo_tags` converts offsets to BIO tags for inspection.


```python
# Verify that entity offsets are correctly aligned to token boundaries
nlp_blank = spacy.blank('es')
text = 'Apple lanzó el nuevo iPhone en California.'
entities = [(0, 5, 'ORG'), (32, 42, 'LOC')]

tags = offsets_to_biluo_tags(nlp_blank.make_doc(text), entities)
print(f'Text: {text}')
print(f'BILUO tags: {tags}')
```

> **Output interpretation:** BILUO tags follow the Begin–Inside–Last–Unit–Outside scheme. `B-ORG` starts the *Apple* entity, `U-LOC` marks *California* as a single-token location. A `-` tag means an offset did not align to a token boundary — misaligned offsets must be fixed before training.


---

## 3. Build Pipeline and Add NER Component


```python
# Start from a blank Spanish model — no pre-trained weights
nlp = spacy.blank('es')

# Add the NER component to the pipeline
ner = nlp.add_pipe('ner')
print(f'Pipeline: {nlp.pipe_names}')
```

---

## 4. Load Annotated Training Data


```python
# TRAIN_DATA format: list of {"text": str, "entities": [(start, end, label), ...]}
with open(DIR_DATA / 'datasetNER.json', 'r', encoding='utf-8') as f:
    train_data = json.load(f)

print(f'Training examples: {len(train_data)}')
print(f'First example: {train_data[0]}')
```

---

## 5. Register Labels and Initialise


```python
# Register all entity labels present in the training data
for entry in train_data:
    for _, _, label in entry['entities']:
        ner.add_label(label)

# Initialise model weights
optimizer = nlp.initialize()
print(f'NER labels: {ner.labels}')
```

---

## 6. Training Loop

Training runs for 30 epochs. The data is shuffled each epoch to reduce ordering bias.


```python
n_epochs = 30

for epoch in range(n_epochs):
    random.shuffle(train_data)
    losses = {}
    for entry in train_data:
        doc = nlp.make_doc(entry['text'])
        example = Example.from_dict(doc, {'entities': entry['entities']})
        nlp.update([example], sgd=optimizer, losses=losses)
    if epoch % 5 == 0:
        print(f'Epoch {epoch:2d}  NER loss: {losses["ner"]:.4f}')
```

> **Output interpretation:** The NER loss decreases as training progresses — the model is learning to predict entity boundaries and labels from the annotated examples. If the loss plateaus or increases, consider adjusting the learning rate, adding more training data, or using a pre-trained base model (`es_core_news_sm`) instead of a blank one.


---

## 7. Save and Reload the Model


```python
model_path = Path('modelo_ner_personalizado')
nlp.to_disk(model_path)
print(f'Model saved to: {model_path.resolve()}')

# Reload and run inference
nlp2 = spacy.load(model_path)
test_text = 'Google abrió una nueva oficina en Madrid.'
doc = nlp2(test_text)

print(f'\nTest text: {test_text}')
print('Entities found:')
for ent in doc.ents:
    print(f'  [{ent.label_}] {ent.text!r}')
```

> **Output interpretation:** The reloaded model (`nlp2`) runs inference on a new sentence. Detected entities show as `[LABEL] text`. The quality depends on the size and diversity of `datasetNER.json`. For production use, evaluate with precision/recall/F1 on a held-out test set using `spacy evaluate` or `evaluate.load('seqeval')`.
