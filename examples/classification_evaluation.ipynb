## 1. Load Models

Install dependencies once: `pip install transformers evaluate`


```python
# pip install transformers evaluate  # run once in terminal
from transformers import (
    AutoTokenizer,
    AutoModelForSeq2SeqLM,
    AutoModelForCausalLM,
    GPT2LMHeadModel,
    GPT2Tokenizer,
)
import evaluate
import math
import torch

T5_MODEL_NAME = 't5-small'
GPT2_MODEL_NAME = 'gpt2'

# Load T5 (encoder–decoder, fine-tuned for seq2seq tasks like summarisation)
t5_tokenizer = AutoTokenizer.from_pretrained(T5_MODEL_NAME)
t5_model = AutoModelForSeq2SeqLM.from_pretrained(T5_MODEL_NAME)

# Load GPT-2 (decoder-only, generative language model)
gpt2_tokenizer = AutoTokenizer.from_pretrained(GPT2_MODEL_NAME)
gpt2_model = AutoModelForCausalLM.from_pretrained(GPT2_MODEL_NAME)

print(f'T5 loaded: {T5_MODEL_NAME}')
print(f'GPT-2 loaded: {GPT2_MODEL_NAME}')
```

> **Output interpretation:** Both models are downloaded from the HuggingFace Hub on first run and cached locally. T5-small has ~60M parameters and is optimised for tasks like summarisation and translation. GPT-2 has ~117M parameters and is optimised for auto-regressive text generation.


---

## 2. Input and Reference Texts


```python
input_text = 'The Eiffel Tower is located in Paris and was built in 1889.'
reference_summary = 'The Eiffel Tower was built in Paris in 1889.'
```

---

## 3. Generate Text with T5


```python
# T5 expects a task prefix; 'summarize:' triggers the summarisation head
t5_input = 'summarize: ' + input_text
inputs_t5 = t5_tokenizer(t5_input, return_tensors='pt')
output_t5 = t5_model.generate(**inputs_t5, max_length=30, temperature=0.7)
text_t5 = t5_tokenizer.decode(output_t5[0], skip_special_tokens=True)
print(f'T5 output: {text_t5}')
```

---

## 4. Generate Text with GPT-2


```python
inputs_gpt2 = gpt2_tokenizer(input_text, return_tensors='pt')
output_gpt2 = gpt2_model.generate(
    **inputs_gpt2,
    max_length=30,
    temperature=0.7,
    do_sample=True,
    top_k=50,
)
text_gpt2 = gpt2_tokenizer.decode(output_gpt2[0], skip_special_tokens=True)
print(f'GPT-2 output: {text_gpt2}')
```

> **Output interpretation:** T5 is fine-tuned for summarisation, so it produces a compressed paraphrase of the input. GPT-2 is a general language model; without task-specific fine-tuning it continues the input sentence rather than summarising it. This difference reflects the importance of **instruction fine-tuning** for task-specific generation.


---

## 5. BLEU and ROUGE Evaluation


```python
bleu = evaluate.load('sacrebleu')
rouge = evaluate.load('rouge')

bleu_t5 = bleu.compute(predictions=[text_t5], references=[[reference_summary]])
rouge_t5 = rouge.compute(predictions=[text_t5], references=[reference_summary])

print(f'Reference:  {reference_summary}')
print(f'T5 output:  {text_t5}')
print(f'BLEU score: {bleu_t5["score"]:.2f}')
print(f'ROUGE-1:    {rouge_t5["rouge1"]:.4f}')
print(f'ROUGE-L:    {rouge_t5["rougeL"]:.4f}')
```

> **Output interpretation:**
>
> - **BLEU** (0–100) measures n-gram precision overlap between prediction and reference. A score of 0 means no n-gram overlap; 100 is a perfect match. For short sentences, BLEU is sensitive to exact phrasing.
> - **ROUGE-1** measures unigram overlap (individual word coverage). **ROUGE-L** measures the longest common subsequence — capturing sentence-level structure.
> - BLEU and ROUGE have well-known limitations: they do not capture semantic equivalence (a perfect paraphrase may score 0) and they favour shorter outputs. More recent metrics like **BERTScore** use embedding similarity to better capture meaning.

## 6. Perplexity (GPT-2)

**Perplexity** measures how well a language model predicts a text. Lower perplexity = better model fit.

PPL(text) = exp(−(1/N) Σ log P(token_i | context))


```python
def compute_perplexity(model: GPT2LMHeadModel, tokenizer: GPT2Tokenizer, text: str) -> float:
    """Compute perplexity of text under a GPT-2 language model.

    Args:
        model: Pre-trained GPT-2 model in eval mode.
        tokenizer: Corresponding tokeniser.
        text: Input string to evaluate.

    Returns:
        Perplexity score (lower = better model fit).
    """
    encodings = tokenizer(text, return_tensors='pt')
    input_ids = encodings.input_ids
    target_ids = input_ids.clone()

    with torch.no_grad():
        outputs = model(input_ids, labels=target_ids)
        neg_log_likelihood = outputs.loss

    return math.exp(neg_log_likelihood.item())


ppl_model = GPT2LMHeadModel.from_pretrained('gpt2')
ppl_tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
ppl_model.eval()

ppl = compute_perplexity(ppl_model, ppl_tokenizer, input_text)
print(f'GPT-2 perplexity on input text: {ppl:.2f}')
```

> **Output interpretation:** Perplexity reflects how *surprised* the model is by the text. A well-formed English sentence about a famous landmark (high-frequency in GPT-2's training data) should yield relatively low perplexity. Higher perplexity indicates the text contains unusual phrasing, domain-specific vocabulary, or language the model was not trained on.

## Metric Summary

| Metric | What it measures | Range | Limitation |
|--------|-----------------|-------|------------|
| **BLEU** | N-gram precision vs. reference | 0–100 | Exact match; no semantics |
| **ROUGE-1** | Unigram recall overlap | 0–1 | Word overlap only |
| **ROUGE-L** | Longest common subsequence | 0–1 | Structural similarity only |
| **Perplexity** | Language model confidence | >1 (lower=better) | Intrinsic; not task-specific |
