```python
# Importamos la librería de OpenAI
import openai

# Establecemos la clave de API de OpenAI para autenticación
openai.api_key = "########"

# Definimos la función para solicitar el prompt al modelo GPT-4o
def generar_prompt_para_tarea():
    # Definimos el prompt para pedirle al modelo GPT-4o que cree el prompt
    #detallado
    prompt = """
    Tu tarea es generar un prompt detallado que será usado para que otro modelo
    genere frases de ejemplo (exemplars) para cada uno de los 21 síntomas del
    Inventario de Depresión de Beck-II (BDI-II) en inglés.

    El prompt generado debe incluir las siguientes condiciones:

    1. Debe generar 20 frases de ejemplo para cada síntoma, escritas en primera
    persona, utilizando pronombres como "I", "me", "my", "myself".
    2. El lenguaje debe ser natural, conversacional y emocionalmente expresivo,
     similar a lo que una persona podría publicar en redes sociales.
    3. El modelo debe devolver el resultado como un diccionario de Python, donde
    cada clave es un número entero del 1 al 21, representando el ID del síntoma,
    y el valor es una lista de 20 cadenas de texto (las frases de ejemplo).
    4. El resultado final debe ser en formato Python válido y no debe incluir
    explicaciones ni formato adicional.

    El prompt debe devolver el resultado como un código Python válido:

    `exemplars_dict = { ... }`
    """

    # Realizamos la solicitud a la API de OpenAI con el modelo GPT-4o
    response = openai.Completion.create(
        model="gpt-4",  # Usar GPT-4 para generar el prompt
        prompt=prompt,
        max_tokens=1000,  # Ajustamos el número de tokens para evitar cortes
        n=1,  # Solo generamos una respuesta
        stop=None,  # No hay tokens de parada específicos
        temperature=0.7  # Controlamos la creatividad de la respuesta
    )

    # Extraemos el código generado del modelo
    prompt_generado = response.choices[0].text.strip()

    # Devolvemos el código generado para su uso posterior
    return prompt_generado

# Llamamos a la función para generar el prompt
prompt_para_tarea = generar_prompt_para_tarea()

# Mostramos el prompt generado para verificar su contenido
print(prompt_para_tarea)
```

```python
# Importar la biblioteca necesaria para interactuar con la API de OpenAI
import openai
import os

# Establecer la clave de API de OpenAI para autenticar las solicitudes
# Reemplazar '###############' con la clave de API correspondiente.


# openai.api_key = "#############"

# Definir una función para generar los ejemplos (exemplars) para los síntomas del BDI-II
def generate_bdi_exemplars():
    # Crear un diccionario vacío donde se almacenarán los resultados
    exemplars_dict = {}

    # Definir el prompt para la solicitud a la API de OpenAI
    prompt = """
    You are a mental health expert helping build a symptom-aware AI model for depression detection.
    Your task is to generate expressive and diverse example sentences (exemplars) in English that reflect each of the 21 symptoms of the Beck Depression Inventory-II (BDI-II).

    Requirements:

    1. Each exemplar must be written in first-person, using pronouns such as "I", "me", "my", or "myself".
    2. Use natural, conversational, and emotionally expressive language as used in real-life social media posts.
    3. Avoid clinical or overly generic wording. Aim for authentic, nuanced expressions.
    4. Each sentence should reflect how a person might naturally express experiencing that symptom.
    5. Return the result as a Python dictionary called `exemplars_dict`, where each key is an integer from 1 to 21 (representing BDI-II symptom ID), and the value is a list of 20 strings (the exemplars).
    6. Output only valid Python code — do not include explanations or extra formatting.

    Here is the list of symptoms with their IDs:

    1. Sadness
    2. Pessimism
    3. Past failure
    4. Loss of pleasure
    5. Guilty feelings
    6. Punishment feelings
    7. Self-dislike
    8. Self-criticalness
    9. Suicidal thoughts or wishes
    10. Crying
    11. Agitation
    12. Loss of interest
    13. Indecisiveness
    14. Worthlessness
    15. Loss of energy
    16. Changes in sleeping pattern
    17. Irritability
    18. Changes in appetite
    19. Concentration difficulty
    20. Tiredness or fatigue
    21. Loss of interest in sex

    Generate the exemplars and assign them directly to `exemplars_dict = { ... }` in valid Python syntax.
    """

    # Realizar la solicitud a la API de OpenAI usando el modelo 'gpt-4'
    response = openai.Completion.create(
        model="gpt-4",  # Usando el modelo más avanzado de OpenAI
        prompt=prompt,
        max_tokens=2000,  # Ajustar según sea necesario para garantizar la respuesta completa
        n=1,  # Solo generar una respuesta
        stop=None,  # No hay un stop explícito
        temperature=0.7,  # Configura la creatividad de la respuesta
    )

    # Extraer la respuesta generada de la API
    generated_code = response.choices[0].text.strip()

    # Ejecutar el código generado directamente
    # Utilizamos eval para que el código sea evaluado como código Python
    exec(generated_code)

    # El diccionario 'exemplars_dict' resultado
    return exemplars_dict

# Llamar a la función para obtener los exemplars
exemplars_dict = generate_bdi_exemplars()

# Ver el resultado final para verificar la estructura
# Mostrar las claves del diccionario para comprobar que se generaron los ejemplos
exemplars_dict.keys()
```
